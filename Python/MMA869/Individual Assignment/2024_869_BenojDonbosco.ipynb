```python
from sklearn.cluster import DBSCAN
dbscan =  DBSCAN(eps = 0.3, min_samples = 3)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
labels = dbscan.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```

```python
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title(f"Estimated number of clusters: {n_clusters_}")
plt.show()
```

## 1.3 Model Comparison

K-means: It is a centroid-based clustering algorithm that aims to partition data into K clusters, where each cluster is represented by its centroid. The algorithm minimizes the sum of squared distances between data points and their assigned centroids.
DBSCAN (Density-Based Spatial Clustering of Applications with Noise): It is a density-based clustering algorithm that groups data points based on their density. It can find clusters of different shapes and handles noise effectively. DBSCAN defines clusters as areas of high density separated by areas of low density.

K-means: It is generally more scalable to large datasets since its time complexity is linear with the number of data points.
DBSCAN: It can be less efficient for high-dimensional and large datasets due to its density-based nature, which requires a larger computation for finding dense regions.

K-means: Clusters are represented by their centroids, making them more interpretable and easier to understand.
DBSCAN: Clusters are defined by connected dense regions, which may not have a straightforward centroid representation, making them less interpretable.

When reviewing the results the clusters produced spherical clusters which Kmeans is better at handling. Kmeans is straightforward to use in comparison to the Dbscan that requires hyperparameters tuning. K-means is easier to interpret as each data point is assigned to the nearest centroid. Therefore I would like to recommend K-means as the preferred clustering algorithm.
